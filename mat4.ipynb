```python
import requests
import math
import sys
try:
    xfile=open("C:/Users/eliot/Desktop/dests.txt",'r',encoding='utf8')
except:
    print("file can not be open")
    sys.exit()
city={}
first_place=int(0)
first_place_name=str()
seconde_place=int(0)
seconde_place_name=str()
third_place=int(0)
third_place_name=str()
for line in xfile:
    line=line.rstrip()
    origins='תל-אביב'
    destinations=line
    try:
        key='Enter your key'
        url="https://maps.googleapis.com/maps/api/distancematrix/json?origins=%s&destinations=%s&key=%s"%(origins,destinations,key)
        url1="https://maps.googleapis.com/maps/api/geocode/json?address=%s&key=%s"%(destinations,key)
        response=requests.get(url).json()
        response1=requests.get(url1).json()
        value_of_time=(response['rows'][0]['elements'][0]['duration']['value'])
        value_of_distance=(response['rows'][0]['elements'][0]['distance']['value'])
    except:
        print("the city does not exist")
        sys.exit()
    if value_of_distance>first_place :
        third_place=seconde_place
        seconde_place=first_place
        first_place=value_of_distance
        third_place_name=seconde_place_name
        seconde_place_name=first_place_name
        first_place_name=line
    if value_of_distance<first_place and value_of_distance>seconde_place  :
        third_place=seconde_place
        seconde_place=value_of_distance
        third_place_name=seconde_place_name
        seconde_place_name=line
    if value_of_distance<seconde_place and value_of_distance>third_place:
        third_place=value_of_distance
        third_place_name=line
    value_of_time_hours=math.trunc(value_of_time/3600)
    value_of_time_minutes=round(((value_of_time/3600)-value_of_time_hours)*60)
    time_to_tlv=str(value_of_time_hours)+' hours '+str(value_of_time_minutes)+' minutes '
    city[line]=((response['rows'][0]['elements'][0]['distance']['text']),time_to_tlv,(response1['results'][0]['geometry']['location']['lat']),(response1['results'][0]['geometry']['location']['lng']))


for key,value in city.items():
    print('Informations for the city: ',key)
    print('Distance to Tel-Aviv: ',value[0])
    print('time to Tel-Aviv: ',value[1])
    print('latitude: ',value[2])
    print('longitude: ',value[3],'\n')

print('the 3 furthest towns from tel aviv are : ',first_place_name,' ',seconde_place_name,' ',third_place_name)

```
